### RandomForest
- 집단 학습을 기반으로 고정밀 분류, 회귀, 클러스터링 등을 구현
- 학습 전용 데이터를 기반으로 다수의 의사결정트리를 만들고 만들어진 의사결정트리를 기반으로 다수 결과를 유도하는 모델
- Ensemble(앙상블) 기법의 지표가 됨 

```python
import pandas as pd
```

```python
iris = pd.read_csv("../Data/iris.csv")
iris.head()
```

```python
# Feature와 Target 분리
iris_data = iris[['SepalLength', 'SepalWidth','PetalLength', 'PetalWidth']]
iris_target = iris['Name']
```

```python
# 확인
print(iris_data.shape)
print(iris_target.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Train과 Test 분리
train_data, test_data, train_target, test_target = train_test_split(
                                                        iris_data,
                                                        iris_target,
                                                        random_state=42,
                                                        stratify=iris_target
)
```

```python
# 확인
print(train_data.shape)
print(test_data.shape)
print(train_target.shape)
print(test_target.shape)
```

```python
# Sampling 판단
print(train_data.describe())
print(test_data.describe())
```

```python
# Target 확인
print(train_target.describe())
print(test_target.describe())
```

---
### RandomForest로 분류

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# 모델 만들기
clf = RandomForestClassifier(n_estimators=1) # Hyper Parameter -> n_estimators : tree 갯수 
```

```python
# 학습시키기
clf.fit(train_data, train_target)
```

```python
# 평가하기
print(clf.score(train_data, train_target))
print(clf.score(test_data, test_target))
```
